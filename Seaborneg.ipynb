```python
!pip install seaborn bokeh

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from bokeh.plotting import figure, show, output_notebook
output_notebook()

sns.set(style="whitegrid")
```

```python

data = pd.DataFrame({
    "day": ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],
    "sales": [100, 120, 90, 140, 160, 180, 200],
    "profit": [20, 30, 15, 40, 50, 60, 70],
    "category": ["A","B","A","B","A","B","A"]
})

data
```

```python
sns.relplot(x="sales", y="profit", data=data, hue="category")
plt.show()
```

```python
sns.relplot(x="sales", y="profit", col="category", data=data)
plt.show()
```

```python
sns.lineplot(x="day", y="sales", data=data)
plt.show()
```

```python
sns.barplot(x="day", y="sales", data=data)
plt.show()
```

```python

plt.figure()
sns.barplot(x="day", y="sales", data=data)
plt.show()

plt.figure()
sns.boxplot(x="category", y="sales", data=data)
plt.show()
```

```python
sns.pointplot(x="day", y="profit", data=data)
plt.show()
```

```python
sns.set_palette("Set3")
sns.barplot(x="day", y="sales", data=data)
plt.show()
```

Bokeh

```python
from bokeh.plotting import figure, show, output_notebook
output_notebook()

p = figure(title="Scatter Plot")

p.scatter(data["sales"], data["profit"], size=10)

show(p)
```

```python
p = figure(title="Patch Plot")

x = [1,2,3,4]
y = [4,6,2,5]

p.patch(x, y)

show(p)
```

```python
p = figure(title="Circle Glyph")

p.circle(data["sales"], data["profit"], size=15)

show(p)
```

```python
p = figure(title="Rectangle")

p.rect(x=[1,2,3], y=[3,4,5], width=0.5, height=1)

show(p)
```

```python
from math import pi

p = figure(title="Wedge")

p.wedge(x=0, y=0, radius=1,
        start_angle=0, end_angle=pi/2)

show(p)
```

```python
p = figure(x_range=(0, 250), y_range=(0, 100),
           title="Custom Range")

p.circle(data["sales"], data["profit"], size=10)

show(p)
```
